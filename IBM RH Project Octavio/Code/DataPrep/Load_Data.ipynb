```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
FilePath = '../../Data/Raw/EmployeeAttrition.csv'


df_original = pd.read_csv(FilePath, sep=',')
df_total = df_original.copy()
df_total.head(2)
```

## Data Description

```python
df_total.describe()
```

# Creating categorical variables

```python
# Attrition

YesAttrition = (df_total['Attrition'] == 'Yes')
NoAttrition = (df_total['Attrition'] == 'No')


df_total.loc[YesAttrition, 'Attrition'] = 1
df_total.loc[NoAttrition, 'Attrition'] = 0
```

```python
# BusinessTravel

df_total['NonTravel'] = 0
df_total['TravelRarely'] = 0
df_total['TravelFrequently'] = 0

Non_Travel = (df_total['BusinessTravel'] == 'Non-Travel')
Trave_Rarely = (df_total['BusinessTravel'] == 'Travel_Rarely')
Travel_Frequently = (df_total['BusinessTravel'] == 'Travel_Frequently')

df_total.loc[Non_Travel, 'NonTravel'] = 1
df_total.loc[Trave_Rarely, 'TravelRarely'] = 1
df_total.loc[Travel_Frequently, 'TravelFrequently'] = 1
```

```python
# Department

df_total['D_R&D'] = 0
df_total['D_Sales'] = 0
df_total['D_RH'] = 0

RD = (df_total['Department'] == 'Research & Development')
Sales = (df_total['Department'] == 'Sales')
RH = (df_total['Department'] == 'Human Resources')

df_total.loc[RD, 'D_R&D'] = 1
df_total.loc[Sales, 'D_Sales'] = 1
df_total.loc[RH, 'D_RH'] = 1
```

```python
# EducationalField

df_total['E_LifeSciences'] = 0
df_total['E_Medical'] = 0
df_total['E_Marketing'] = 0
df_total['E_TechnicalDegree'] = 0
df_total['E_RH'] = 0
df_total['E_Other'] = 0

LifeSciences = (df_total['EducationField'] == 'Life Sciences')
Medical = (df_total['EducationField'] == 'Medical')
Marketing = (df_total['EducationField'] == 'Marketing')
TechnicalDegree = (df_total['EducationField'] == 'Technical Degree')
RH = (df_total['EducationField'] == 'Human Resources')
Other = (df_total['EducationField'] == 'Other')

df_total.loc[LifeSciences, 'E_LifeSciences'] = 1
df_total.loc[Medical, 'E_Medical'] = 1
df_total.loc[Marketing, 'E_Marketing'] = 1
df_total.loc[TechnicalDegree, 'E_TechnicalDegree'] = 1
df_total.loc[RH, 'E_RH'] = 1
df_total.loc[Other, 'E_Other'] = 1
```

```python
# Gender

df_total['Male'] = 0
df_total['Female'] = 0

Male = (df_total['Gender'] == 'Male')
Female = (df_total['Gender'] == 'Female')

df_total.loc[Male, 'Male'] = 1
df_total.loc[Female, 'Female'] = 1
```

```python
# marital Status

df_total['Single'] = 0
df_total['Married'] = 0
df_total['Divorced'] = 0

Single = (df_total['MaritalStatus'] == 'Single')
Married = (df_total['MaritalStatus'] == 'Married')
Divorced = (df_total['MaritalStatus'] == 'Divorced')

df_total.loc[Single, 'Single'] = 1
df_total.loc[Married, 'Married'] = 1
df_total.loc[Divorced, 'Divorced'] = 1
```

```python
# OverTime



YesOverTime = (df_total['OverTime'] == 'Yes')
NoOverTime = (df_total['OverTime'] == 'No')

df_total.loc[YesOverTime, 'OverTime'] = 1
df_total.loc[NoOverTime, 'OverTime'] = 0
```

```python
# JobRole

df_total['JR_LaboratoryTechnician'] = 0
df_total['JR_SalesRepresentative'] = 0
df_total['JR_ResearchScientist'] = 0
df_total['JR_RH'] = 0
df_total['JR_ManufacturingDirector'] = 0
df_total['JR_SalesExecutive'] = 0
df_total['JR_HealthcareRepresentative'] = 0
df_total['JR_Research Director'] = 0
df_total['JR_Manager'] = 0

LaboratoryTechnician = (df_total['JobRole'] == 'Laboratory Technician')
SalesRepresentative = (df_total['JobRole'] == 'Sales Representative')
ResearchScientist = (df_total['JobRole'] == 'Research Scientist')
RH = (df_total['JobRole'] == 'Human Resources')
ManufacturingDirector = (df_total['JobRole'] == 'Manufacturing Director')
SalesExecutive = (df_total['JobRole'] == 'Sales Executive')
HealthcareRepresentative = (df_total['JobRole'] == 'Healthcare Representative')
ResearchDirector = (df_total['JobRole'] == 'Research Director')
Manager = (df_total['JobRole'] == 'Manager')


df_total.loc[LaboratoryTechnician, 'JR_LaboratoryTechnician'] = 1
df_total.loc[SalesRepresentative, 'JR_SalesRepresentative'] = 1
df_total.loc[ResearchScientist, 'JR_ResearchScientist'] = 1
df_total.loc[RH, 'JR_RH'] = 1
df_total.loc[ManufacturingDirector, 'JR_ManufacturingDirector'] = 1
df_total.loc[SalesExecutive, 'JR_SalesExecutive'] = 1
df_total.loc[HealthcareRepresentative, 'JR_HealthcareRepresentative'] = 1
df_total.loc[ResearchDirector, 'JR_Research Director'] = 1
df_total.loc[Manager, 'JR_Manager'] = 1
```
