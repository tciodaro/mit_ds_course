# Data Preparation
### This notebook prepare data to be consumed by model.

```python
import pandas as pd
from ipynb.fs.full.lib_data_analiser import *
```

```python
data = pd.read_parquet('../data/raw/dataFromFcApi.parquet')
data
```

```python
# descartando colunas que não servem
data = data.drop(['t'], axis=1)
data = data.drop(['v'], axis=1)
```

```python
data.columns = ["open", "hight", "low", "close", "time"]
data.head(5)
```

```python
print("Total de linhas: {}".format(data.shape[0]))
quantificarDadosFaltantes(data)
```

```python
# convertendo dados para numérico
numerics = data.copy().drop(['time'], axis=1)
for col in numerics.columns:
    numerics[col] = pd.to_numeric(numerics[col])
    data[col] = pd.to_numeric(data[col])
```

```python
# convertendo coluna data
data['time'] = pd.to_datetime(data['time'])
```

```python
# sorting and working with last events:
data = data.sort_values('time')
data = data.tail(10000)
data
```

```python
# descartando variáveis:
dataAux = data.drop(['open', 'hight', 'low'], axis=1)
```

```python
# criando features

columns = ['close']
intervals = [1, 2, 3, 5, 10, 15, 20, 25, 30, 40, 50, 60, 90, 120, 240, 360, 720, 1440, 2880, 4320]

for x in intervals:

    moved = dataAux[columns].shift(-x)
    moved.ffill(axis ='rows', inplace=True)
    
    if x/60 < 2:
        moved.columns = [str(x) + ' min']
    elif x/1440 < 1:
        moved.columns = [str(int(x/60)) + ' hours']
    else:
        moved.columns = [str(int(x/1440)) + ' days']
    
    dataAux = pd.concat([moved, dataAux], axis=1)

dataAux
```

```python
quantificarDadosFaltantes(dataAux)
```
