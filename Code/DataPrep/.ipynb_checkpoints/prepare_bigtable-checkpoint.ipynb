```python
import pandas as pd


data_directory = '../../Data/Processed/wine/'
bigtable_path = '../../Data/Processed/wine/bigtable_data.parquet'
```

# Load Data 

```python
red_wine_file = data_directory + '/red.parquet'
white_wine_file    = data_directory + '/white.parquet'

# read features
df_red = pandas.read_parquet(red_wine_file)
df_white = pandas.read_parquet(white_wine_file)

print('red wine:', df_red.shape)
print('white wine:', df_white.shape)
```

# Combine the Data Sets 

```python
df_red['type'] = 0
df_red.head()
```

```python
df_white['type'] = 1
df_white.head()
```

```python
df = df_red.append(df_white)
print('data shape:', df.shape)
print('data columns:', df.columns.tolist())
df.head()
```

```python
df.corr()
```

# Faixa Dinamica

```python
variacao = df.max()-df.min()
variacao.head()
```

```python
faixa_dinamica = pd.concat([df.min(),df.max(),variacao], axis=1)
faixa_dinamica.columns = ["Min", "Max", "Diff"]
faixa_dinamica
```

# Export Big Table Data 

```python
df.to_parquet(bigtable_path)
```
