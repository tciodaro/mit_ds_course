```python
import pandas
import os


data_directory = '../../Data/Raw/wine/'
output_directory='../../Data/Processed/wine/'
```

# Load Data 

```python
red_wine_file   = data_directory + '/winequality-red.csv'
white_wine_file = data_directory + '/winequality-white.csv'

# read red wine
df_red = pandas.read_csv(red_wine_file, sep=';')
print('Red Wine: ', df_red.shape)

# read white wine
df_white = pandas.read_csv(white_wine_file, sep=';')
print('White Wine: ', df_white.shape)
```

# Check for NaN Values 

```python
dados_faltantes_red = df_red.isnull()
dados_faltantes_red.head()
```

```python
dados_faltantes_red.sum()/dados_faltantes_red.shape[0]
```

```python
dados_faltantes_white = df_white.isnull()
dados_faltantes_white.head()
```

```python
dados_faltantes_white.sum()/dados_faltantes_white.shape[0]
```

# Numeric Variables 

```python
# Red Wine Data
num_variables = ['fixed acidity', 'volatile acidity', 'citric acid','residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide',
                 'density', 'pH', 'sulphates', 'alcohol']

for cname in num_variables:
    df_red[cname] = df_red[cname].astype(float, errors='ignore').fillna(0)

    
# White Wine Data    

for cname in num_variables:
    df_white[cname] = df_white[cname].astype(float, errors='ignore').fillna(0)
    
    

```

# Export Data

```python
help(os.makedirs)
```

```python
os.makedirs(output_directory, exist_ok=True)

red_wine_file = output_directory + '/red.parquet'
white_wine_file = output_directory + '/white.parquet'

# read features
df_red.to_parquet(red_wine_file)
df_white.to_parquet(white_wine_file)
```
