# Data Collect
### This notebook collect data from FC API

```python
import pandas as pd
import fastparquet
from datetime import datetime
from ipynb.fs.full.lib_forex_fc_api_helper import *
```

```python
# Reading pre existing data file
data = pd.read_parquet('../data/raw/dataFromFcApi.parquet')
print(data.shape[0], ' elements')
data.head(5)
```

```python
# Recovering date time from first and last event:
firstEvent = data['tm'].min()
lastEvent = data['tm'].max()
print('We have data from ', firstEvent, ' to ', lastEvent)
```

```python
# Recovering data from last update until now:
print('Recovering data from ' + lastEvent + ' until now.')
#newData = getDataFromInterval("EUR/USD", "1m", "01-08-2020 00:00", "05-08-2020 00:00")
newData = getDataFromInterval("EUR/USD", "1m", lastEvent, None)
newData
```

```python
# Appending new data:
data = data.append(newData)
print(data.shape[0])
data.head(5)
```

```python
# Saving new collection:
data.to_parquet('../data/raw/dataFromFcApi.parquet', compression='GZIP')
```
