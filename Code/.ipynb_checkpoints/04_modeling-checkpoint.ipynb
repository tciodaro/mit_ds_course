```python
#!pip install fbprophet
import pandas as pd
import pickle
from fbprophet import Prophet
from datetime import timedelta
from datetime import datetime
from ipynb.fs.full.lib_data_analiser import *

minutesReservedForTest = 30
modelFile = '../data/model/trainedModel.parquet'
```

```python
# Loading data:
data = pd.read_parquet('../data/processed/dataProcessed.parquet')
data
```

```python
# Recovering date time from first and last event:
firstEvent = data['time'].min()
lastEvent = data['time'].max()
print('We have data from ', firstEvent, ' to ', lastEvent)
```

```python
# Renaming columns for forecast:
data = data.rename(columns={'close':'y', 'time':'ds'})
data.head(5)
```

```python
# Training model:

data = data.tail(1000)

treino = data[data['ds'] < (data['ds'].max() - timedelta(minutes=minutesReservedForTest))]
teste = data[data['ds'] >= (data['ds'].max() - timedelta(minutes=minutesReservedForTest))]

model = Prophet()
model.fit(treino)

print('Model trained!')
```

```python
# Predicting:
future = model.make_future_dataframe(periods=30, include_history=True, freq='min')
forecast = model.predict(future)
print('Predict ok!')
```

```python
print(forecast.shape[0])
forecast
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```

```python
# Separating predicted data:
print('Predicted until ', treino['ds'].max())
predicted = forecast[forecast['ds'] > treino['ds'].max()]
```

```python
# Ploting results:

pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt

plt.figure(figsize=[5,3])
plt.plot(teste['ds'], teste['y'])
plt.plot(predicted['ds'], predicted['yhat'])

#teste.plot(x='ds', y='y', label='Real')
#predicted.plot(x='ds', y='yhat', label='Forecast')

plt.legend(['Real','Forecast'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid()
```

```python
# Score model
lastPredictedValue = predicted.tail(1)['yhat'].values[0]
lastRealValue = teste.tail(1)['y'].values[0]
print('Last predicted=', lastPredictedValue, 'Last real=', lastRealValue)
score = mape(lastRealValue, lastPredictedValue)
print(' mape %.2f %%'%score)
```

```python
# Refiting with all data
model = Prophet()
model.fit(data)
print('Model retrained!')
```

```python
# Saving model
with open(modelFile, 'wb') as fid:
    pickle.dump(model, fid)
```
