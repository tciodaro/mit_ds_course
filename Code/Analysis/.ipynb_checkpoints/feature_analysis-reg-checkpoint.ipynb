Nessa etapa vamos prever as idades dos passageiros para preencher os dados faltantes e com isso conseguir fazer uma predição mais assertiva se esse passageiro sobreviveu ou não ao desastre.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')

datapath = '../../Data/Processed/titanic.parquet'
save_file = '../../Data/Processed/titanic_reg_model.parquet'
```

# Leitura dos Dados 

```python
df = pd.read_parquet(datapath)

print('shape:', df.shape)
print('columns:', df.columns)
```

# Analisando o Dataframe

```python
df.describe()
```

```python
df.describe(include=['O'])
```

# Extraindo o título de cada passageiro da variável Nome.

Como vimos anteriormente, a variável Nome é do tipo Alfanumérica e pode conter erros de abreviação e possuir títulos. 
Com isso vamos extrair os títulos da variável Nome

```python
df['Titulo'] = df.Nome.str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
pd.crosstab(df['Titulo'], df['Sexo'])
```

Podemos ver que existem muitas abreviações de títulos, o que pode afetar o nosso modelo. Vamos então reduzir esse dados.

```python
df['Titulo'] = df['Titulo'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Outros')

df['Titulo'] = df['Titulo'].replace('Mlle', 'Miss')
df['Titulo'] = df['Titulo'].replace('Ms', 'Miss')
df['Titulo'] = df['Titulo'].replace('Mme', 'Mrs')

pd.crosstab(df['Titulo'], df['Sexo'])
```

# Verficando a média de sobreviventes de acordo com os títulos

```python
df[['Titulo', 'Sobreviveu']].groupby(['Titulo'], as_index=False).mean()
```

Com isso podemos verificar que mulheres e crianças tem uma maior chance de sobrevivência que os homens adultos. Isso mostra a importancia da variável Idade
para a construção do nosso modelo.

# Traduzindo os títulos

```python
traduzindo_titulos = {
    'Master': 'Menino',
    'Miss': 'Solteira',
    'Mr': 'Adulto',
    'Mrs': 'Casada',
    'Outros': 'Outros'
    
}

df['Titulo'] = df['Titulo'].map(traduzindo_titulos)

pd.crosstab(df['Titulo'], df['Sexo'])
```

# Verificando se podemos melhorar a coluna Outros

```python
df.loc[(df.Titulo=='Outros')]
```

```python
df.loc[(df.Titulo=='Outros')].describe()['Idade']
```

* Podemos ver que nenhum passageiro nessa categoria é menor que 18 anos, tendo uma média de idade de 45 anos. 
* Apesar de ter uma idade faltante, esse passageiro não viaja com mais ninguém, então podemos assumir que ele não é uma criança.
* Portanto, não temos nenhuma criança nessa categoria.
* Também podemos ver que existe dentre as mulheres, uma que viaja com o marido. Então podemos atribuir a ela o Titulo = Casada.

## Alterando o título dos passageiros

```python
for idx, _ in df.iterrows():
    if idx in df.loc[(df.Titulo=='Outros')].index:
        if (df['Sexo'][idx] == 'Homem'):
            df['Titulo'].at[idx] = 'Adulto'
        elif (df['Sexo'][idx] == 'Mulher') & (df['IrmaosCasal'][idx] == 1):
            df['Titulo'].at[idx] = 'Casada'
        else:
            df['Titulo'].at[idx] = 'Solteira'
```

```python
pd.crosstab(df['Titulo'], df['Sexo'])
```

# Verificando as informações do DataFrame

Com os dados já tratatdos podemos então remover essas colunas.

```python
df.drop(['Nome','Bilhete'], axis = 1, inplace = True)
```

Vamos analisar novamente o Crosstab de Título e Idade.

```python
pd.crosstab(df['Titulo'], df['Sexo'])
```

Conseguimos identificar com precisão quais são os meninos e homens adultos, porém não conseguimos diferenciar as meninas das mulheres adultas solteiras. 
Precisamos refinar nossa informação fazendo uma análise pelas idades e se viajam ou não acompanhada dos pais.  
  
Vamos analisar então cada situação.

# Solteira viajando com os pais

Vamos analisar a media das mulheres solteiras que estão a bordo.

### Média das solteiras a bordo.

```python
df['Idade'].loc[(df['Titulo'] == 'Solteira')].mean()
```

### Analisando as solteiras que viajam com os pais

```python
solteiras_pais = df.loc[(df['Titulo'] == 'Solteira') & (df['PaisFilhos'] >= 1)]
solteiras_pais.head()
```

```python
solteiras_pais['Idade'].mean()
```

Podemos ver que a média de idade já diminuiu bastante. Vamos analisar com gráficos.

```python
plt.hist(solteiras_pais.loc[solteiras_pais['Titulo']=='Solteira']['Idade'], alpha=0.5, label = 'Solteira com pais');
plt.hist(df.loc[df['Titulo']=='Solteira']['Idade'], alpha=0.5, label = 'Solteira' );
plt.legend(loc='upper right');
plt.show();
```

### Média das mulheres casadas

```python
plt.hist(df.loc[df['Titulo']=='Casada']['Idade'], bins = 15);
```

Podemos concluir que a maioria das mulheres solteiras que não viajam com os pais tem uma idade mais avançada, mais parecida com a idade das mulheres casadas.      
Vamos adicionar essa coluna de SolteiraPais na tabela para melhorar nosso modelo iterando pelo index para adicionar na posição correta.

```python
df['SolteiraPais'] = 0

for idx, _ in df.iterrows():
    if idx in solteiras_pais.index:
        df['SolteiraPais'].at[idx] = 1
        
df.head()
```

Vamos calculara novamente a média das solteiras, com base se elas viajaram ou não com os pais.

### Média solteira sem pais x Média solteira com pais

```python
print(' A média das solteiras sem pais é %.2f' %  df.loc[df.SolteiraPais == 0]['Idade'].mean())
print(' A média das solteiras com pais é %.2f' %  df.loc[df.SolteiraPais == 1]['Idade'].mean())
print(' A média das casadas é %.2f' %  df.loc[df.Titulo == 'Casada']['Idade'].mean())
```

```python
# Colocando a classificacao nas tarifas

df.loc[df['Tarifa']<102,'Tarifa']=0
df.loc[(df['Tarifa']>102) & (df['Tarifa']<204),'Tarifa']=1
df.loc[(df['Tarifa']>204) & (df['Tarifa']<307),'Tarifa']=2
df.loc[df['Tarifa']>307, 'Tarifa']=3
```

```python
df.Tarifa.unique()
```

# Verificando a correlação

```python
sns.heatmap(df.corr(), annot = True, cmap='Blues');
```

É possível observar que a correlação entre a Idade (variável-alvo) e Sobreviveu é bem baixa (-0.077).  
Como essa variável contém muitos dados nulos, iremos excluir para uma melhor construção do nosso modelo.

```python
df.drop('Sobreviveu', axis = 1, inplace = True)
df.isnull().sum()
```

Somente a variável Idade está como nula, e é justamente ela que querermos prever.  

Precisamos agora transformar nossas variáveis categóricas em dummy para conseguirmos treinar nosso modelo.

```python
df.head()
```

```python
df['Sexo']=df['Sexo'].map({'Homem':0,'Mulher':1})
```

## Transformando em dummys as variáveis Sexo, Classe, Embarque e Título.

```python
df = pd.get_dummies(df, columns=['Classe', 'PortoEmbarque','Titulo'], drop_first=True)
df.head()
```

# Exportanto Base de Dados

```python
df.to_parquet(save_file)
```
